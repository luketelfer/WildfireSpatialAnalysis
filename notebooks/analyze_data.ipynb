```python
import numpy as np
import pandas as pd
import pylandstats as pls
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import rasterio
import glob
import seaborn as sns
sns.set(rc={'figure.figsize':(10,5)})
```

# Load Data

```python
lmetrics = pd.read_csv('../metric_data/landscape_metrics.csv')
cmetrics = pd.read_csv('../metric_data/class_metrics.csv')
classes = ['unburned','low severity','moderate severity','high severity']
```

# Exploration

```python
proportion = cmetrics[['burn_severity_class','proportion_of_landscape']]
proportion = proportion[proportion['burn_severity_class'].isin(classes)]
sns.boxplot(x="burn_severity_class",y="proportion_of_landscape",data=proportion)
plt.show()
```

```python
patchDensity = cmetrics[['burn_severity_class','patch_density']]
patchDensity = patchDensity[patchDensity['burn_severity_class'].isin(classes)]
sns.boxplot(x="burn_severity_class",y="patch_density",data=patchDensity)
plt.show()
```

```python
LPI = cmetrics[['burn_severity_class','largest_patch_index']]
LPI = LPI[LPI['burn_severity_class'].isin(classes)]
sns.boxplot(x="burn_severity_class",y="largest_patch_index",data=LPI)
plt.show()
```

```python
cmetrics = cmetrics[cmetrics['burn_severity_class'].isin(classes)]
sns.pairplot(cmetrics, hue="burn_severity_class", height=2.5, kind='reg');
```

```python
l = sns.pairplot(lmetrics, height=2.5, diag_kind='kde', kind='reg');
plt.show()
```
